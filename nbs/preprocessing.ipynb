## Rename imgs and labels to have the same name

```python
folder = "/Users/gregruyoga/Code/gmoneycodes/scrape_me_plate/data/imgs/yolov8/valid"
```

```python
for filename in os.listdir(f"{folder}/images"):
    if filename.endswith(".jpg"):
        parts = filename.split(".")
        if len(parts) >= 4:
            new_name = f"{parts[2]}.jpg"
            old_file = os.path.join(f"{folder}/images", filename)
            new_file = os.path.join(f"{folder}/images", new_name)
            
            os.rename(old_file, new_file)
            print(f"Renamed {filename} to {new_name}")
        else:
            print(f"Skipping {filename} cus its not in a valid format")
    else:
        print(f"Skipping {filename} cus its not a jpeg")
```

```python
for filename in os.listdir(f"{folder}/labels"):
    if filename.endswith(".txt"):
        parts = filename.split(".")
        if len(parts) >= 4:
            new_name = f"{parts[2]}.txt"
            old_file = os.path.join(f"{folder}/labels", filename)
            new_file = os.path.join(f"{folder}/labels", new_name)

            os.rename(old_file, new_file)
            print(f"Renamed {filename} to {new_name}")
        else:
            print(f"Skipping {filename} cus its not in a valid format")
    else:
        print(f"Skipping {filename} cus its not a jpeg")
```

```python
def bounding_box(image_path):
    label_path = image_path.replace('/images/', '/labels/')
    label_path = label_path.replace('.jpg', '.txt')

    image = Image.open(image_path)
    draw = ImageDraw.Draw(image)
    
    with open(label_path, 'r') as f:
        for line in f:
            label, x, y, w, h = line.split(' ')
            
            x = float(x)
            y = float(y)
            w = float(w)
            h = float(h)

            # Convert center position, width, height into
            # top-left and bottom-right coordinates
            W, H = image.size
            print(image.size)
            x1 = (x - w/2) * W
            y1 = (y - h/2) * H
            x2 = (x + w/2) * W
            y2 = (y + h/2) * H

            # Draw the bounding box with red lines
            draw.rectangle((x1, y1, x2, y2),
                           outline=(255, 0, 0), 
                           width=5)      
    image.show()
            
```
